```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([5, 54])
ypoints = np.array([4, 34])

plt.title('Simple line plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(xpoints, ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([5,7,10,34,54])
ypoints = np.array([4,6,23,29, 34])

plt.title('Simple line plot with circle points and green lines')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(xpoints, ypoints, 'o:g')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([5,7,10,1,20])
ypoints = np.array([4,6,9,2,13])
colours=['red','pink','yellow','green','blue']

plt.title('Simple scatter plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.scatter(xpoints, ypoints,color=colours)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 15]
colours=['red','pink','yellow','green']


plt.title('Bar Graph using Seaborn')
plt.xlabel('Categories')
plt.ylabel('Values')
sns.barplot(x=categories, y=values)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

values = [1, 1, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 6]


plt.title('Histogram using Seaborn')
plt.xlabel('Frequency')
plt.ylabel('Values')
sns.histplot(values)
plt.show()
```
