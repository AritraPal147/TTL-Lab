```python
# Python Libraries
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading the dataset and displaying the top 5 records with Dataframe.head() function
data=pd.read_csv("data.csv")
data.shape
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
plt.plot(data['Year'], data['GDP Growth'])

# Customize the graph
plt.xlabel('Year')
plt.ylabel('GDP Growth Rate')
plt.xticks(rotation = 75)
plt.title('GDP Growth Rate over Time(Yearly Data)')
plt.grid(True)

# Display the graph
plt.show()
```

```python
data.set_index('Year', inplace = True)
data['Recession'] = ((data['GDP Growth'] < 0) & (data['GDP Growth'].shift(1) < 0))

# Fill missing values with False (since the first quarter cannot be in a recession)
data['Recession'].fillna(False, inplace=True)
```

```python
data.head()
```

```python
# Identify the time period where recession happen.
# recession_data = data.loc[data['Recession'] == True]
# print(recession_data)
plt.plot(data['Year'], data['GDP Growth'])

# Plot recession periods
# plt.scatter(recession_data['Year'], recession_data['GDP Growth'], color='Red', label='Recession')

# Customize the plot
plt.xlabel('Year')
plt.ylabel('GDP Growth')
plt.xticks(rotation=75)
plt.title('GDP Growth and Recession Over Time')
plt.legend()

# Display the plot
plt.show()
```
