```python
# Python Libraries
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Reading the dataset and storing it in a dataframe
data=pd.read_csv("data.csv")
```

```python
# Printing the top 5 rows of the data
data.head()
```

```python
# Sets figure size for graph to 10 x 6 inches
plt.figure(figsize=(10, 6))  

# Plotting the Year vs GDP Growth Graph
plt.plot(data['Year'], data['GDP Growth'])

# Customize the graph by labelling
plt.xlabel('Year')
plt.ylabel('GDP Growth Rate')
plt.title('GDP Growth Rate over Time(Yearly Data)')
plt.grid(True)

# Display the graph
plt.show()
```

```python
# Creating a new dictionary to store old data + recession data
dict = { 'Year':[], 'GDP Growth': [], 'Recession':[]}    

# newData is the dataframe format of dict
newData = pd.DataFrame(dict) 

# Copying Year and GDP Growth from data variable to newData variable
newData['Year'] = data['Year']
newData['GDP Growth'] = data['GDP Growth']

# Filling the Recession column by subtracting the previous year's GDP from the current's GDP
# Note: Recession occurs when year-over-year growth is negative, i.e., when GDP of current 
# year - GDP of last year is negative.
newData["Recession"] = data['GDP Growth'] - data['GDP Growth'].shift(1) < 0

# Printing top 5 rows of the new data
newData.head()
```

```python
# Identify the years where recession happen.
# recessionData variable contains the years in which recession has occured and the 
# corresponding GDP growth
recessionData = newData.loc[newData["Recession"] == True]

# Sets figure size for graph to 10 x 6 inches
plt.figure(figsize=(10, 6))  
# Plotting the Year vs GDP Growth Graph
plt.plot(data['Year'], data['GDP Growth'], label="GDP Growth")

# Plot recession periods
plt.scatter(recessionData['Year'], recessionData['GDP Growth'], color='Red', label='Recession')

# Customize the plot
plt.xlabel('Year')
plt.ylabel('GDP Growth')
plt.title('GDP Growth and Recession Over Time')
plt.legend()
plt.grid()

# Display the plot
plt.show()
```
